### Visualizing the data

```python
fig = plt.figure(figsize = (40,20))
# creating the bar plot
plt.bar(df['YEAR'],df['FUEL CONSUMPTION'] , color ='maroon',width=0.7)
 
plt.xlabel("YEAR")
plt.ylabel("FULE CONSUMPTION")
plt.title("FULE CONSUMPTION PER VEHICLE YEAR")
plt.show()
```

```python
sns.histplot(df.EMISSIONS)
```

```python
sns.histplot(df['FUEL CONSUMPTION'])
```

```python
sns.scatterplot(data=df['YEAR'], x=df['HWY (L/100 km)'], y=df['COMB (L/100 km)'])
```

```python
df['MODEL'].value_counts().head(10).plot(kind='pie',autopct="%0.1f%%").set_title('Top Ten Car Model')
```

```python
df['VEHICLE CLASS'].value_counts().head(5).plot(kind='pie',autopct="%0.1f%%").set_title('Top Five VEHICLE CLASS')
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
```

```python
sns.catplot(data=df,x='FUEL',y='EMISSIONS',kind='bar',col='YEAR',col_wrap=5,sharex=False)
plt.xticks(rotation='vertical')
```


A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

```python
sns.kdeplot(data=df, x="EMISSIONS")
# continuous probability density curve of emissions
```
